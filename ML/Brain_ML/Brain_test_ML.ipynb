# Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Charger, observer et comprendre le dataset

```python
df = pd.read_csv('brain_tumor_dataset.csv')
df.head()
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

# Transformer les données catégorielles en données numériques

```python
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder

ord_encod = OrdinalEncoder()
lab_encod = LabelEncoder()

features = df.columns
features = features.drop("MRI_Result")
features = features.drop("Patient_ID")

X = df[features]
y = df["MRI_Result"]

X = X.copy()
X_categorielles = X.select_dtypes(include=['object', 'category']).columns
X[X_categorielles] = ord_encod.fit_transform(X[X_categorielles])
y = lab_encod.fit_transform(y)

y_visu = pd.DataFrame(y, columns=['MRI_Results'])
X.head()
```

```python
y_visu.head()
```

# Standardiser les données

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(X)

X_visu = pd.DataFrame(X, columns=features)
X_visu.head(10)
```

# Séparer le dataset en train et test

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y)
```

# Régression logistique
Entrainer une régression logistique avec les hyper-paramètres par défaut

```python
from sklearn.linear_model import LogisticRegression

log = LogisticRegression()
log.fit(X_train, y_train)

#Sans le predict
print(f'Scoring of default {log.__class__} train: {log.score(X_train,y_train)}')
print(f'Scoring of default {log.__class__} test : {log.score(X_test,y_test)}')

#Avec le predict
y_train_pred = log.predict(X_train)
y_test_pred = log.predict(X_test)

# Comparer avec les valeurs réelles
accuracy_train = (y_train_pred == y_train).mean()
accuracy_test = (y_test_pred == y_test).mean()
print(f"Accuracy on test set: {accuracy_train}")
print(f"Accuracy on test set: {accuracy_test}")

```

# KNN
Entrainer un KNN avec les hyper-paramètres par défaut
Avec `GridSearchCV` ou `RandomizedSearchCV`, optimiser les hyper-paramètres. Utiliser une validation croisée de 5 splits. Vous explorerez les paramètres suivants:
- k : 3, 5, 7, 9, 11,
- poids : uniform, distance,
- distances : euclidean, manhattan, minkowski

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)

print(f'Scoring of default {knn.__class__} train: {knn.score(X_train,y_train)}')
print(f'Scoring of default {knn.__class__} test : {knn.score(X_test,y_test)}')
```

```python
from sklearn.model_selection import GridSearchCV

knn = KNeighborsClassifier()
param_grid = {
    'n_neighbors': [3, 5, 7, 9, 11],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan', 'minkowski']
}
grid_search_knn = GridSearchCV(knn, param_grid, cv=5)
grid_search_knn.fit(X_train, y_train)

best_knn = grid_search_knn.best_estimator_

print(f'Scoring of GridSearchCV {best_knn.__class__} train: {best_knn.score(X_train,y_train)}')
print(f'Scoring of GridSearchCV {best_knn.__class__} test : {best_knn.score(X_test,y_test)}')
print(f'Best parameters: {grid_search_knn.best_params_}')
```

# SVM

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.svm import SVC

param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
    'gamma': ['scale', 'auto', 0.001, 0.01, 0.1, 1],
    'degree': [2, 3, 4, 5]
}

rand_search = RandomizedSearchCV(SVC(), param_grid, cv=5, n_iter=20)

rand_search.fit(X_train, y_train)

best_svm = rand_search.best_estimator_

print(f'Scoring of GridSearchCV {best_svm.__class__} train: {best_svm.score(X_train,y_train)}')
print(f'Scoring of GridSearchCV {best_svm.__class__} test : {best_svm.score(X_test,y_test)}')
print(f'Best parameters: {rand_search.best_params_}')
```
